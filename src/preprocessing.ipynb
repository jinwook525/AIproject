```python
import pandas as pd

# 원본 데이터 로드
file_path = '../document/data_after_dropping.csv'
data = pd.read_csv(file_path)

# 시간 데이터 변환
data['timestamp'] = pd.to_datetime(data['timestamp'])

# 시간 순 정렬
data = data.sort_values(['mmsi', 'timestamp'])

# MMSI별로 결측치와 이전/이후 값 확인
result = []

for mmsi, group in data.groupby('mmsi'):
    group = group.reset_index(drop=True)  # 인덱스 재설정
    for i, row in group.iterrows():
        if pd.isna(row['lat']) or pd.isna(row['lon']):  # 위도 또는 경도가 NaN인 경우
            # 이전 값 확인
            has_previous_lat = False
            has_previous_lon = False

            if i > 0:  # 첫 번째 행이 아닌 경우
                has_previous_lat = not pd.isna(group.loc[:i - 1, 'lat']).all()
                has_previous_lon = not pd.isna(group.loc[:i - 1, 'lon']).all()

            # 이후 값 확인 (현재 행 이후 모든 행에서 값이 존재하는지 확인)
            has_next_lat = not pd.isna(group.loc[i + 1:, 'lat']).all()
            has_next_lon = not pd.isna(group.loc[i + 1:, 'lon']).all()

            # 결과 저장
            result.append({
                'mmsi': mmsi,
                'timestamp': row['timestamp'],
                'lat_is_nan': pd.isna(row['lat']),
                'lon_is_nan': pd.isna(row['lon']),
                'has_previous_lat': has_previous_lat,
                'has_previous_lon': has_previous_lon,
                'has_next_lat': has_next_lat,
                'has_next_lon': has_next_lon
            })

# 결과 데이터프레임 생성
result_df = pd.DataFrame(result)

# 결측치 처리: 조건에 맞는 데이터 제외
filtered_result = result_df[
    ~(
        (result_df['lat_is_nan'] & result_df['lon_is_nan']) &  # 현재 lat/lon이 NaN이고
        (~result_df['has_previous_lat'] & ~result_df['has_previous_lon']) &  # 이전 값도 없고
        (~result_df['has_next_lat'] & ~result_df['has_next_lon'])  # 이후 값도 없는 경우
    )
]

# 필터링된 데이터에서 'mmsi'와 'timestamp'만 사용하여 원본 데이터와 병합
valid_data = filtered_result[['mmsi', 'timestamp']]

# 원본 데이터와 유효 데이터 병합하여 결측치 제거
final_data = pd.merge(data, valid_data, on=['mmsi', 'timestamp'], how='inner')

# 최종 데이터 저장
final_data.to_csv('../document/final_cleaned_data.csv', index=False, encoding='utf-8-sig')

print("결측치가 제거된 최종 데이터가 'final_cleaned_data.csv' 파일로 저장되었습니다.")
```

```python
import pandas as pd

# 원본 데이터 로드
file_path = '../document/data_after_dropping.csv'
data = pd.read_csv(file_path)

# 시간 데이터 변환
data['timestamp'] = pd.to_datetime(data['timestamp'])

# 시간 순 정렬
data = data.sort_values(['mmsi', 'timestamp'])

# MMSI별 결측치와 이전/이후 값 확인
result = []

for mmsi, group in data.groupby('mmsi'):
    group = group.reset_index(drop=True)
    for i, row in group.iterrows():
        if pd.isna(row['lat']) or pd.isna(row['lon']):  # 위도 또는 경도가 NaN인 경우
            # 이전 값 확인
            has_previous_lat = False
            has_previous_lon = False

            if i > 0:
                has_previous_lat = not pd.isna(group.loc[:i - 1, 'lat']).all()
                has_previous_lon = not pd.isna(group.loc[:i - 1, 'lon']).all()

            # 이후 값 확인
            has_next_lat = not pd.isna(group.loc[i + 1:, 'lat']).all()
            has_next_lon = not pd.isna(group.loc[i + 1:, 'lon']).all()

            # 결과 저장
            result.append({
                'mmsi': mmsi,
                'timestamp': row['timestamp'],
                'lat_is_nan': pd.isna(row['lat']),
                'lon_is_nan': pd.isna(row['lon']),
                'has_previous_lat': has_previous_lat,
                'has_previous_lon': has_previous_lon,
                'has_next_lat': has_next_lat,
                'has_next_lon': has_next_lon
            })

# 결과 데이터프레임 생성
result_df = pd.DataFrame(result)

# 결측치 처리 조건에 맞는 데이터 제외
to_remove = result_df[
    (result_df['lat_is_nan'] & result_df['lon_is_nan']) &  # 현재 lat/lon이 NaN이고
    (~result_df['has_previous_lat'] & ~result_df['has_previous_lon']) &  # 이전 값도 없고
    (~result_df['has_next_lat'] & ~result_df['has_next_lon'])  # 이후 값도 없는 경우
]

# 제거할 데이터 확인
print(f"제거할 데이터 행 수: {len(to_remove)}")

# 원본 데이터에서 제거할 데이터 제외
cleaned_data = data[~data.set_index(['mmsi', 'timestamp']).index.isin(to_remove.set_index(['mmsi', 'timestamp']).index)]

# 최종 데이터 저장
cleaned_data.to_csv('../document/final_cleaned_data.csv', index=False, encoding='utf-8-sig')

print(f"최종 데이터가 저장되었습니다. 최종 데이터 행 수: {len(cleaned_data)}")
```
